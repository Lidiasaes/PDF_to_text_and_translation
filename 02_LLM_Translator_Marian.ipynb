```python
from transformers import MarianMTModel, MarianTokenizer
from langdetect import detect
import os

target_language = 'fr' # set target language
def load_txt_files(input_file_path):
    with open(input_file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
    content = ''.join(lines)
    return content

def detect_language(content):
    try:
        # Detect the language, if supported
        lang = detect(content)
        return lang
    except:
        # In case of any error, default to English ('en')
        return 'en'

def translate_sentence(sentence, source_lang, target_lang=target_language):
    model_name = f'Helsinki-NLP/opus-mt-{source_lang}-{target_lang}'
    model = MarianMTModel.from_pretrained(model_name)
    tokenizer = MarianTokenizer.from_pretrained(model_name)

    input_ids = tokenizer.encode(sentence, return_tensors="pt")
    translated_tokens = model.generate(input_ids, max_length=512)
    translated_sentence = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)

    return translated_sentence

def translate_content_and_print_sentences(content, source_lang, target_lang=target_language):
    # Split content into sentences. Here, a simple split by period is used.
    # For more complex scenarios, consider using nltk or another library that can accurately split text into sentences.
    sentences = content.split('.')
    for sentence in sentences:
        if sentence.strip() == "":  # Skip empty sentences
            continue
        translated_sentence = translate_sentence(sentence, source_lang, target_lang)
        print("Original sentence:", sentence.strip())
        print("Translated sentence:", translated_sentence.strip())
        print("---")
```

```python
# the .txt file we have obtained from pdf (script 01)
input_folder_path = '/content/drive/My Drive/Colab Notebooks/Translator_LLM/input_data'

for file_name in os.listdir(input_folder_path):
    if file_name.endswith(".txt"):
        file_path = os.path.join(input_folder_path, file_name)
        content = load_txt_files(file_path)
        source_lang = detect_language(content)  # Detect once per document
        print('Source language detected:', source_lang)
        translate_content_and_print_sentences(content, source_lang)
        print("Finished translation for", file_name)
```
